```python
spark.sql("""
CREATE OR REPLACE TEMP VIEW Curated.matches AS
SELECT
  CAST(match_id AS STRING)                        AS match_id,
  CAST(TO_DATE(match_date, 'yyyy-MM-dd') AS DATE) AS match_date,
  CAST(player_id AS INT)                           AS player_id,
  CAST(runs AS INT)                                AS runs,
  CAST(balls AS INT)                               AS balls,
  CAST(ROUND((runs / balls) * 100, 2) AS DOUBLE)   AS strike_rate,
  CAST(fours AS INT)                               AS fours,
  CAST(sixes AS INT)                               AS sixes,
  CAST((fours * 4) + (sixes * 6) AS INT)           AS boundary_runs,
  CAST(wickets AS INT)                             AS wickets,
  CAST(catches AS INT)                             AS catches,
  CAST(CURRENT_TIMESTAMP() AS TIMESTAMP)           AS load_ts
FROM (
  SELECT *, ROW_NUMBER() OVER (PARTITION BY match_id, player_id ORDER BY match_date) AS rn
  FROM matches
  WHERE match_id IS NOT NULL
    AND player_id IS NOT NULL
)
WHERE rn = 1
""")
```
