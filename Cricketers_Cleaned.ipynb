```python
spark.sql("""
CREATE OR REPLACE TEMP VIEW Curated.cricketers AS
SELECT
  CAST(player_id AS INT)                AS player_id,
  CAST(INITCAP(TRIM(Player_name)) AS STRING)  AS player_name,
  CAST(
    CASE
      WHEN LOWER(TRIM(Country)) = 'austraila' THEN 'Australia'
      ELSE INITCAP(TRIM(Country))
    END AS STRING
  )                                     AS country,
  CAST(INITCAP(TRIM(Role)) AS STRING)   AS role,
  CAST(CURRENT_TIMESTAMP() AS TIMESTAMP) AS load_ts
FROM (
  SELECT *, ROW_NUMBER() OVER (PARTITION BY player_id ORDER BY player_id) AS rn
  FROM cricketers
  WHERE player_id IS NOT NULL
    AND TRIM(Player_name) != ''
)
WHERE rn = 1
""")
```
